Создайте матрицу 3 на 3, заполненную числами от 1 до 9
Прибавьте к ней матрицу с единицами на диагонале
Умножьте полученную матрицу на вектор (1, 2, 3)
Посчитайте обратную матрицу к полученной матрице

```python
import numpy as np
```

```python
M = np.arange(9).reshape((3,3))
print(M)
M = M + np.eye(3)
print(M)
M = M * np.array([1,2,3])
print(M)
M = np.linalg.inv(M)
print(M)
```

Уровень 1:
Дан объект в 2d пространстве:

При помощи матричных операций выполните:
Уменьшить объект в два раза.
Повернуть на 130 градусов.
Отразить объект относительно прямой y=x
Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) и масштабирование на (0.5, 1,2)

```python
import matplotlib.pyplot as plt
from math import cos, sin, pi
```

```python
A = [[-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]]
A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Уменьшить объект в два раза.

```python
T = [[.5,0,0],
     [0,.5,0],
     [0,0,1]]

T = np.array(T)
TA = A @ T
print(TA)
plt.plot(A[:,0], A[:,1])
plt.plot(TA[:,0], TA[:,1])
plt.legend(['Оригинальный', 'Уменьшенный'])

plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Повернуть на 130 градусов. ( по часовой стрелке)

```python
N_grad = 130
N_rad = N_grad * pi/180
T =[[cos(N_rad), -sin(N_rad), 0],
    [sin(N_rad), cos(N_rad), 0],
    [0, 0, 1]]

T = np.array(T)
TA = A @ T

plt.plot(A[:,0], A[:,1])
plt.plot(TA[:,0], TA[:,1])
plt.legend(['Оригинальный', 'Повернутый на 130'])

plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Отразить объект относительно прямой y=x

```python
T = [[0,1,0],
     [1,0,0],
     [0,0,1]]

T = np.array(T)
TA = A @ T

plt.plot(A[:,0], A[:,1])
plt.plot(TA[:,0], TA[:,1])
plt.legend(['Оригинальный', 'Отраженный'])
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) и масштабирование на (0.5, 1,2)

```python
T1 = [[.5,0,0],
     [0,1.2,0],
     [0,0,1]]

T2 = [[1,0,0],
     [0,1,0],
     [200,300,1]]

T1 = np.array(T1)
T2 = np.array(T2)

TA = A @ T1 @ T2   

plt.plot(A[:,0], A[:,1])
plt.plot(TA[:,0], TA[:,1])
plt.legend(['Оригинальный', 'Перенесенный+масштаб'])
plt.ylim([-200, 450])
plt.xlim([-200, 300])
```
